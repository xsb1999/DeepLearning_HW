```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list_mse_train_1 = [1.283, 1.279]
list_mse_test_1 = [1.070, 1.008]
```

### 原始和考虑月份（探究月份对负荷有无影响（不考虑小时））

```python
x = ['原始','考虑月份']
plt.figure(figsize=(8, 5))
plt.plot(x,list_mse_train_1,'b-*',label = 'train')
plt.plot(x,list_mse_test_1, 'r-*',label = 'test')
plt.ylabel('MSE')
plt.legend(loc='center right')
# plt.xticks([])
plt.show
```

### 原始和考虑小时（探究小时对负荷有无影响（不考虑月份））

```python
list_mse_train_2 = [1.283, 0.904, 0.912, 0.907, 0.981]
list_mse_test_2 = [1.070, 0.980, 0.981, 0.973, 1.013]
```

```python
x = ['原始','法1','法2','法3','法4']
plt.figure(figsize=(8, 5))
plt.plot(x,list_mse_train_2,'b*-',label = 'train')
plt.plot(x,list_mse_test_2, 'r*-',label = 'test')
plt.ylabel('MSE')
plt.legend(loc='best')
# plt.xticks([])
plt.show
```

### 探究小时对负荷的影响（在考虑月份的条件下）

```python
list_mse_train_3 = [1.283, 0.942, 0.920, 0.857, 0.954]
list_mse_test_3 = [1.070, 0.902, 0.945, 0.935, 0.978]
```

```python
x = ['原始','法1','法2','法3','法4']
plt.figure(figsize=(8, 5))
plt.plot(x,list_mse_train_3,'b*-',label = 'train')
plt.plot(x,list_mse_test_3, 'r*-',label = 'test')
plt.ylabel('MSE')
plt.legend(loc='best')
# plt.xticks([])
plt.show
```

### 探究考虑月份与不考虑月份的差别（train）（在考虑小时的条件下）

```python
x = ['原始','法1','法2','法3','法4']
plt.figure(figsize=(8, 5))
plt.plot(x,list_mse_train_3,'b*-',label = 'with month')
plt.plot(x,list_mse_train_2, 'r*-',label = 'without month')
plt.ylabel('MSE')
plt.legend(loc='best')
# plt.xticks([])
plt.show
```

### 探究考虑月份与不考虑月份的差别（test）（在考虑小时的条件下）

```python
x = ['原始','法1','法2','法3','法4']
plt.figure(figsize=(8, 5))
plt.plot(x,list_mse_test_3,'b*-',label = 'with month')
plt.plot(x,list_mse_test_2, 'r*-',label = 'without month')
plt.ylabel('MSE')
plt.legend(loc='best')
# plt.xticks([])
plt.show
```
