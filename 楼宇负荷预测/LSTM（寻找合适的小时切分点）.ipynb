## 不同小时对用电的影响

```python
df = pd.read_csv("data/dataset_with_hours.csv")
```

```python
df
```

```python
df = df.drop(['Unnamed: 0', 'timestamp'], axis=1)
```

```python
hours = df['hour']
df_x = df.drop(['hour'], axis=1)
```

```python
hours
```

```python
df_x
```

```python
x = df_x.mean(axis=1)
```

```python
xx = pd.DataFrame({ 'x' :x, 'hour' :hours})
xx
```

```python
res = []
```

```python
for i in range(24):
    res.append(xx['x'][xx['hour']==i].mean())
```

```python
y = np.arange(24)
y
```

```python
res
```

```python
hour_df = pd.DataFrame({'hour' :y,'average_Electricity_load' :res})
hour_df
```

```python
from matplotlib.pyplot import MultipleLocator
plt.figure(figsize=(10, 6))
ax=plt.gca()
x_major_locator=MultipleLocator(1)
ax.xaxis.set_major_locator(x_major_locator)
plt.xlim(0,23)
plt.axvline(0.5, color='g')
plt.axvline(10.5, color='g')
plt.axvline(13.5, color='g')
plt.axvline(21.5, color='g')
plt.axhline(np.array(res).mean(), color='blue', linestyle='--')
plt.plot(y, res,'r*')
```

```python
from matplotlib.pyplot import MultipleLocator
plt.figure(figsize=(12, 7.5))
ax=plt.gca()
x_major_locator=MultipleLocator(1)
ax.xaxis.set_major_locator(x_major_locator)
plt.xlim(0,23)
plt.axvline(0.5, color='g',linestyle='--')
plt.axvline(10.5, color='g',linestyle='--')
plt.axvline(13.5, color='g',linestyle='--')
plt.axvline(21.5, color='g',linestyle='--')
plt.axhline(np.array(res).mean(), color='blue', linestyle='--')
plt.plot(y, res,'ro-')
```

## 不同月份对用电的影响

```python
df1 = pd.read_csv("data/data_with_month.csv")
```

```python
df1 = df1.drop(['Unnamed: 0', 'timestamp'], axis=1)
df1
```

```python
month = df1['month']
df1_x = df1.drop(['month'], axis=1)
```

```python
x1 = df1_x.mean(axis=1)
x1
```

```python
xx1 = pd.DataFrame({ 'x' :x1, 'month' :month})
xx1
```

```python
res1 = []
for i in range(6,10):
    res1.append(xx1['x'][xx1['month']==i].mean())
```

```python
y1 = np.arange(6,10)
y1
```

```python
month_df = pd.DataFrame({'month' :y1,'average_Electricity_load' :res1})
month_df
```

```python
from matplotlib.pyplot import MultipleLocator
plt.figure(figsize=(10, 6))
ax=plt.gca()
x_major_locator=MultipleLocator(1)
ax.xaxis.set_major_locator(x_major_locator)
# plt.xlim(6,9)
plt.plot(y1, res1,'rs--')
```
